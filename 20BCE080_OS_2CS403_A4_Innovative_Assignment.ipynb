```python
# firstFit function
def firstFit(List_Blocks,Number_Of_Block,List_Process, Number_Of_Process):
    print("***************************************************First Fit Algorithm*********************************************************")
    print("\n")
    allocation = [-1] * Number_Of_Process
    for i in range(Number_Of_Process):
        for j in range(Number_Of_Block):
            if List_Blocks[j] >= List_Process[i]:
                allocation[i] = j
                List_Blocks[j] -= List_Process[i]
                print("Block Configuration After "+str(i+1)+" Iteration is : ",end="")
                if(allocation[i]!=-1):
                    for k in range(Number_Of_Block):
                        if(k==allocation[i]):
                            print("p" + str(i+1) + "|" + str(List_Blocks[k]),end=" ")
                        else:
                            print(List_Blocks[k],end=" ")
                print("\n")
                print("Remaing Hole size After "+str(i+1)+" Iteration is : ",end="")
                print("P" + str(i+1) + " | At Block No. " + str(j+1) + " | Remaining space : "+ str(List_Blocks[j]),end="   ")
                print("\n")
                print("\n")
                break
    print("\n")
    print("*******************************************************************************************************************************")
    print("                                                        OUTPUT                                                                          ")
    print("*******************************************************************************************************************************")
         
    print("\t\t\tProcess No.\t\t\tProcess Size\t\t\tBlock no.")
    print("*******************************************************************************************************************************")
    for i in range(Number_Of_Process):
        print("\t\t\t", i + 1, "\t\t\t\t", List_Process[i],
                          "\t\t\t\t", end = " ")
        if allocation[i] != -1:
            print(allocation[i] + 1)
        else:
            print("Not Allocated")
    print("*******************************************************************************************************************************")

    
# nextFit function
def nextFit(List_Blocks,Number_Of_Block,List_Process,  Number_Of_Process):
    print("***************************************************Next Fit Algorithm*********************************************************")
    print("\n")
    allocation = [-1] * Number_Of_Process
    j=0
    for i in range(Number_Of_Process):
        while j < Number_Of_Block:
            if List_Blocks[j] >= List_Process[i]:
                allocation[i] = j
                List_Blocks[j] -= List_Process[i]
                print("Block Configuration After "+str(i+1)+" Iteration is : ",end="")
                if(allocation[i]!=-1):
                    for k in range(Number_Of_Block):
                        if(k==allocation[i]):
                            print("p" + str(i+1) + "|" + str(List_Blocks[k]),end=" ")
                        else:
                            print(List_Blocks[k],end=" ")
                print("\n")
                print("Remaing Hole size After "+str(i+1)+" Iteration is : ",end="")
                print("P" + str(i+1) + " | At Block No. " + str(j+1) + " | Remaining space : "+ str(List_Blocks[j]),end="   ")
                print("\n")
                print("\n")
                break
            j = (j + 1) % Number_Of_Block
    print("\n")
    print("*******************************************************************************************************************************")
    print("                                                        OUTPUT                                                                          ")
    print("*******************************************************************************************************************************")
         
    print("\t\t\tProcess No.\t\t\tProcess Size\t\t\tBlock no.")
    print("*******************************************************************************************************************************")
    for i in range(Number_Of_Process):
        print("\t\t\t", i + 1, "\t\t\t\t", List_Process[i],
                          "\t\t\t\t", end = " ")
        if allocation[i] != -1:
            print(allocation[i] + 1)
        else:
            print("Not Allocated")
    print("*******************************************************************************************************************************")


# bestFit function
def bestFit(List_Blocks,Number_Of_Block,List_Process,  Number_Of_Process):
    print("***************************************************Best Fit Algorithm*********************************************************")
    print("\n")
    allocation = [-1] * Number_Of_Process
    for i in range(Number_Of_Process):
        bestIdx = -1
        for j in range(Number_Of_Block):
            if List_Blocks[j] >= List_Process[i]:
                if bestIdx == -1:
                    bestIdx = j
                elif List_Blocks[bestIdx] > List_Blocks[j]:
                    bestIdx = j
        if bestIdx != -1:
            allocation[i] = bestIdx
            List_Blocks[bestIdx] -= List_Process[i]
            print("Block Configuration After "+str(i+1)+" Iteration is : ",end="")
            if(allocation[i]!=-1):
                for k in range(Number_Of_Block):
                    if(k==allocation[i]):
                        print("p" + str(i+1) + "|" + str(List_Blocks[k]),end=" ")
                    else:
                        print(List_Blocks[k],end=" ")
            print("\n")
            print("Remaing Hole size After "+str(i+1)+" Iteration is : ",end="")
            print("P" + str(i+1) + " | At Block No. " + str(allocation[i]+1) + " | Remaining space : "+ str(List_Blocks[bestIdx]),end="   ")
            print("\n")
            print("\n")
    print("\n")
    print("*******************************************************************************************************************************")
    print("                                                        OUTPUT                                                                          ")
    print("*******************************************************************************************************************************")
         
    print("\t\t\tProcess No.\t\t\tProcess Size\t\t\tBlock no.")
    print("*******************************************************************************************************************************")
    for i in range(Number_Of_Process):
        print("\t\t\t", i + 1, "\t\t\t\t", List_Process[i],
                          "\t\t\t\t", end = " ")
        if allocation[i] != -1:
            print(allocation[i] + 1)
        else:
            print("Not Allocated")
    print("*******************************************************************************************************************************")


# worstFit function
def worstFit(List_Blocks,Number_Of_Block,List_Process,  Number_Of_Process):
    print("***************************************************Worst Fit Algorithm*********************************************************")
    print("\n")
    allocation = [-1] * Number_Of_Process
    for i in range(Number_Of_Process):
        wstIdx = -1
        for j in range(Number_Of_Block):
            if List_Blocks[j] >= List_Process[i]:
                if wstIdx == -1:
                    wstIdx = j
                elif List_Blocks[wstIdx] < List_Blocks[j]:
                    wstIdx = j
        if wstIdx != -1:
            allocation[i] = wstIdx
            List_Blocks[wstIdx] -= List_Process[i]
            print("Block Configuration After "+str(i+1)+" Iteration is : ",end="")
            if(allocation[i]!=-1):
                for k in range(Number_Of_Block):
                    if(k==allocation[i]):
                        print("p" + str(i+1) + "|" + str(List_Blocks[k]),end=" ")
                    else:
                        print(List_Blocks[k],end=" ")
            print("\n")
            print("Remaing Hole size After "+str(i+1)+" Iteration is : ",end="")
            print("P" + str(i+1) + " | At Block No. " + str(allocation[i]+1) + " | Remaining space : "+ str(List_Blocks[wstIdx]),end="   ")
            print("\n")
            print("\n")
        else:
            print("Remaing Hole size After "+str(i+1)+" Iteration is : ",end="")
            print("P" + str(i+1) + " | Not Allocated " + " | Remaining space : ---",end="   ")
            print("\n")
            print("\n")
    print("\n")
    print("*******************************************************************************************************************************")
    print("                                                        OUTPUT                                                                          ")
    print("*******************************************************************************************************************************")
         
    print("\t\t\tProcess No.\t\t\tProcess Size\t\t\tBlock no.")
    print("*******************************************************************************************************************************")
    for i in range(Number_Of_Process):
        print("\t\t\t", i + 1, "\t\t\t\t", List_Process[i],
                          "\t\t\t\t", end = " ")
        if allocation[i] != -1:
            print(allocation[i] + 1)
        else:
            print("Not Allocated")
    print("*******************************************************************************************************************************")



# Asking for total number of Blocks And Process
print("**********************************************NUMBER OF BLOCKS AND PROCESSES***************************************************")
Number_Of_Block   =  int(input("Number Of Blocks  : "))
Number_Of_Process =  int(input("Number of Process : "))
print("*******************************************************************************************************************************")
print("\n")

# making list to store size of blocks and process
List_Blocks=[];List_Process=[]
    
# Asking for size of each Blocks And Process
print("***************************************************SIZE OF BLOCKS**************************************************************")
for i in range(0,Number_Of_Block):
    x=int(input("size of block " + str(i+1) + " : "))
    List_Blocks.append(x)
print("*******************************************************************************************************************************")
print("\n")
print("***************************************************SIZE OF PROCESSES***********************************************************")
for i in range(0,Number_Of_Process):
    y=int(input("size of process " + str(i+1) + " : "))
    List_Process.append(y)
print("*******************************************************************************************************************************")
print("\n")

# ask for choice
print("*******************************************************************************************************************************")
print("Enter your choice:")
print("Enter 1 to perform first fit algorithm")
print("Enter 2 to perform next fit algorithm")
print("Enter 3 to perform best fit algorithm")
print("Enter 4 to perform worst fit algorithm")
print("Enter E to Exit from the process")
choice=(input("Choice: "))
print("*******************************************************************************************************************************")
print("\n")

if(choice=='1'):
    firstFit(List_Blocks,Number_Of_Block,List_Process,  Number_Of_Process)
elif(choice=='2'):
    nextFit(List_Blocks,Number_Of_Block,List_Process,  Number_Of_Process)
elif(choice=='3'):
    bestFit(List_Blocks,Number_Of_Block,List_Process,  Number_Of_Process)
elif(choice=='4'):
    worstFit(List_Blocks,Number_Of_Block,List_Process,  Number_Of_Process)
elif(choice=='E'):
    print("You Are Successfully Exit. Thank You...")
    exit()
else:
    print("Sorry! Invalid Choice. You Are Automatically Exit")
    
    
    
# ****************************************************************************************************************************
# Example
# blockSize = [100, 600, 200, 300, 700]
# processSize = [118, 350, 156, 499] 
# ***************************************************************************************************************************
```
